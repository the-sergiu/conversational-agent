### Suno (large)
- Does not support Romanian
- Absolutely amazing quality for English text, but has to be chunked in speech bits which would take at most 13-14 seconds to speak. As such, long-form text has to be broken up into multiple chunks.

```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline, AutoProcessor, AutoModel
from datasets import load_dataset

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(device)
```

```python
voice_preset = 'v2/pt_speaker_0'
processor = AutoProcessor.from_pretrained("suno/bark", voice_preset=voice_preset)
model = AutoModel.from_pretrained("suno/bark").to(device)
```

```python
inputs = processor(
    # text=["Hello, my name is Suno. And, uh — and I like pizza. [laughs] But I also have other interests such as playing tic tac toe."],
    text=["Bună! Numele meu este Suno și sunt agentul dumneavoastră virtual! Cu ce vă pot ajuta?"],
    return_tensors="pt",
)

speech_values = model.generate(**inputs.to(device), do_sample=True)
```

```python
from IPython.display import Audio

sampling_rate = model.generation_config.sample_rate
Audio(speech_values.cpu().numpy().squeeze(), rate=sampling_rate)
```

```python
import scipy

sampling_rate = model.generation_config.sample_rate
scipy.io.wavfile.write("bark_out.wav", rate=sampling_rate, data=speech_values.cpu().numpy().squeeze())
```
